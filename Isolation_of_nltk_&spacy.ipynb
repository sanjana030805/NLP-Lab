```python

input = "Hi I am Sanjana . how are you"
output1 = word_tokenize(input)
print(output1)
```

```python

output2 = sent_tokenize(input)
print(output2)
```

```python

import spacy
nlp = spacy.load("en_core_web_sm")
input = " This is the second experiment"
doc = nlp(input)
```

```python

for token in doc:
  print(token)
```

```python

for sent in doc.sents:
  print(sent)
```
